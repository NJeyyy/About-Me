````python
import  numpy  as  np
import pandas  as  pd
import lightgbm as lgb
from sklearn.model_selection  import train_test_split
#from sklearn.preprocessing  import LabelBinarizer
#from datetime import datetime
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
````

````python
"""
Using this just to hide unnecessary stuff (atleast for me)
"""
from contextlib import contextmanager
import sys, os
@contextmanager
def suppress_stdout():
    with open(os.devnull, "w") as devnull:
        old_stdout = sys.stdout
        sys.stdout = devnull
        try:
            yield
        finally:
            sys.stdout = old_stdout
````

````python
Ds = pd.read_csv("sensor_data (2).csv", parse_dates=["Timestamp"])
sensors_dataset = ["Humidity", "Temperature", "Soil Moisture", "MQ135 Value"]
Ds = Ds.sort_values("Timestamp")
# Create time-based features
Ds['hour'] = Ds['Timestamp'].dt.hour
Ds.dropna(inplace=True)
print(Ds)
````

````python
def split_train_lgb(dts, target_name):
    shifted_target = dts[target_name].shift(-1)
    shifted_target.dropna(inplace=True)
    ft = dts[["hour", target_name]].iloc[:-1]
    xr, xs, yr, ys = train_test_split(ft, shifted_target, test_size=0.3, shuffle=False)
    tr_data = lgb.Dataset(xr, label=yr)
    ts_data = lgb.Dataset(xs, label=ys)
    params = {
        'objective': 'regression',
        'metric': 'rmse',
        'boosting_type': 'gbdt',
        'num_leaves': 31,
        'learning_rate': 0.05,
        'feature_fraction': 0.9
    }

    num_round = 100
    with suppress_stdout():
      model = lgb.train(params, tr_data, num_round, valid_sets=[tr_data, ts_data])
    y_pred = model.predict(xs, num_iteration=model.best_iteration)
    # Calculate RMSE
    rmse = mean_squared_error(ys, y_pred, squared=False)
    #return model
    return {"model": model, "rmse": rmse}
````

````python
plcm = {}
for w in sensors_dataset:
  plcm[w] = split_train_lgb(Ds, w)

print("{")
for i in plcm:
  print(" ", i, ": {\n  model:", plcm[i]["model"], "\n  RMSE:", plcm[i]["rmse"], "\n }")
print("}")
print(plcm)
````

````python
import matplotlib.pyplot as plt
import numpy as np

# define data values
x = Ds["hour"]  # X-axis points
y_htsm = {}
for i in plcm:
  y_htsm[i] = plcm[i]["model"].predict(Ds[["hour", i]])  # Y-axis points


for r in y_htsm:
  print(r)
  plt.plot(x, y_htsm[r])  # Plot the chart
  plt.show()  # display
````
